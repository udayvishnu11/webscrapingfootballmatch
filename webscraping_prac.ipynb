```python
import requests,json

url = 'http://www.collegehockeystats.net/1920/teamstats/brnm'
r = requests.get(url)

print(len(r.text))
table = r.text.split("table>")[1]
lines = table.split('\n')
print(lines)
n = 0
rc = 0
cc = 0
d = {}
for line in lines:
    if "<strong" in line:
        rc+=1
    if "<td" in line:
        cc+=1
    if "</strong>" in line:
        cc = 0
    if cc == 2 and rc == 1:
        rh = line.split(">")[1].split("<")[0]
        d['rh'] = rh

    if cc == 2 and rc == 2:
        wind = line.split(">")[1].split("<")[0]
        d['wind'] = wind
print(d)
```
